```python
import pandas as pd
```

```python
df_start = pd.read_csv('trans_clean.csv')
df_start.head()
```

```python
len(df_start)
```

```python
df = df_start.sort_values(by = ['bond_sym_id', 'trd_exctn_dt'])
```

```python
def detect_outlier(detect_df):
    detect_df['prev_diff_perc'] = detect_df['rptd_pr'].pct_change()
    detect_df['next_diff_perc'] = detect_df['rptd_pr'].pct_change(-1)
    detect_df['is_outlier'] = (((detect_df['prev_diff_perc'] > 0.5) & (detect_df['next_diff_perc'] > 0.5))|((detect_df['prev_diff_perc'] < -0.5) & (detect_df['next_diff_perc'] <-0.5)))
    return detect_df
df = df.groupby('bond_sym_id',group_keys=False).apply(detect_outlier)
df.head()
```

```python
df_true = df[df['is_outlier'] == True]
df_true.head()
```

```python
df_true['bond_sym_id'].nunique()
```

```python
len(df_true)
```

```python
df_false = df[df['is_outlier'] == False]
df_false.head()
```

```python
df_false['bond_sym_id'].nunique()
```

```python
len(df_false)
```

```python
df_final = df_start[df_start['bond_id'].isin(df_false['bond_id'])]
```

```python
df_final.head()
```

```python
len(df_final)
```

```python
df_final.set_index('bond_id', inplace=True)
```

```python
df_final.to_csv('trans_clean_outlier.csv')
```
